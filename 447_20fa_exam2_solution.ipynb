# problem (1.)

Consider the nonlinear system 
$$ \dot{x_1} = -x_1 + x_1^3,\ \dot{x_2} = -2x_2, $$
which has 3 equilibria:  $(0,0)$, $(1,0)$, $(-1,0)$.

## subproblem (a.)

Linearize the system at one of the equilibria.

**Solution:**\
In state-space form, the nonlinear system is
$$\dot{x} = \frac{d}{dt} \left[\begin{matrix} x_1 \\ x_2 \end{matrix}\right] = \left[\begin{matrix} -x_1 + x_1^3 \\ -2x_2 \end{matrix}\right] = f(x).$$

The Jacobian derivative of $f$ with respect to $x$ is
$$ \partial_x f = \left[\begin{matrix} \partial_{x_1} \dot{x}_1 &  \partial_{x_2} \dot{x}_1 \\  \partial_{x_1} \dot{x}_2 &  \partial_{x_2} \dot{x}_2 \end{matrix}\right] = \left[\begin{matrix} 3x_1^2 - 1 & 0 \\ 0 & -2 \end{matrix}\right].$$

Evaluating the Jacobian at $x_{e1} = (0, 0)$ yields
$$ A_1 = \partial_x f(x_{e1}) = \boxed{\left[\begin{matrix} -1 & 0 \\ 0 & -2 \end{matrix}\right]}$$

Evaluating the Jacobian at $x_{e2} = (1, 0)$ yields
$$ A_2 = \partial_x f(x_{e2}) = \boxed{\left[\begin{matrix} 2 & 0 \\ 0 & -2 \end{matrix}\right]}$$

Evaluating the Jacobian at $x_{e3} = (-1, 0)$ yields
$$ A_3 = \partial_x f(x_{e3}) = \boxed{\left[\begin{matrix} 2 & 0 \\ 0 & -2 \end{matrix}\right]}$$

## subproblem (b.)

Determine whether the equilibrium chosen in (a.) is stable.
**Solution:**\
For equilibria $x_{e1}$: $(0,0)$,\
The eigenvalue of $A_1$ is $(-1, -2)$, the system is stable.

For equilibria $x_{e2}$: $(1,0)$,\
The eigenvalue of $A_1$ is $(2, -2)$, the system is unstable.

For equilibria $x_{e3}$: $(-1,0)$,\
The eigenvalue of $A_1$ is $(2, -2)$, the system is unstable.

# problem (2.)

A process model $\dot{y} = u$ relating scalar input $u$ to scalar output $y$ has state-space representation

$$ \dot{x} = a x + b u,\ y = c x + d u $$

where $a = 0$, $b = 1$, $c = 1$, $d = 0$.  

An observer-based controller for the process has state-space representation

$$ \dot{\widehat{x}} = a \widehat{x} + b u - \ell(y - \widehat{y}),\ \widehat{y} = c \widehat{x} + d u,\ u = -k\widehat{x} $$

where $k,\ell\in\mathbb{R}$ are the controller's parameters.

## subproblem (a.)

Determine the transfer functions of the process, $G_{yu}(s)$, and the controller, $G_{uy}(s)$.

**Solution:**\
For the transfer function of the process,\
$G_{yu}(s)=c(s-a)^{-1}b+d = \boxed{\frac{1}{s}}$

In order to determine the transfer function $G_{uy}$, we need to write the system in to:\
$\dot{\widehat{x}} = a'\widehat{x}+b'y$\
$u=c'\widehat{x}+d'y$\
From the question, we have:\
$\dot{\widehat{x}} = a \widehat{x} + b u - \ell(y - \widehat{y})$,\
plug in the values and substitute $u$ and $\widehat{y}$:\
$\dot{\widehat{x}} = -k\widehat{x} + \ell (y-\widehat{x}) = (\ell - k)\widehat{x} - \ell y$,\
And we have $u = -k\widehat{x}$\
thus, we have $a' = \ell - k$, $b' = -\ell$, $c' = -k$, $d' = 0$\
And the trainsfer fuction:\
$G_{uy}= c'(s-a')^{-1}b' + d' = \boxed{\frac{k\ell}{s+k-\ell}}$


## subproblem (b.)

Verify that the closed-loop system obtained by interconnecting the process and controller through $y$ and $u$ is stable for any $k > 0$, $\ell < 0$.

**Solution:**\
The system is stable when $Re\lambda (a-bk) < 0$ and $Re\lambda (a + \ell c) < 0$\
Thus, $a - bk<0 \rightarrow \boxed{k>0}$,\
$a + \ell c<0 \rightarrow \boxed{\ell<0}$

# problem (3.)

Consider $L(s) = P(s) C(s) = 20 / (s + 1)^3$, the ***(open-)loop transfer function*** for a control system.

## subproblem (a.)

Create the ***Nyquist diagram*** and determine whether the closed-loop system is stable using the ***Nyquist stability criterion***.

**Solution:**

According to the following Nyquist diagram, it can be seen that the critical point is encircled, ***and the open-loop transfer function has no poles in the right half-plane***. Thus, the Nyquist stability criterion ensures the system is unstable.


```
import numpy as np
import matplotlib.pyplot as plt
```

```
# code for 3a.
L = lambda s,a=1 : 20. / (s + a)**3

# equivalently:

def L(s,a=1):
  return 20. / (s + a)**3

omega = np.linspace(0.,1e1,1000)

Omega = L(1.j*omega)

plt.figure()
ax = plt.subplot(1,1,1); ax.grid('on')
ax.plot(Omega.real,Omega.imag,'b-',label=r'$L(j\omega), \omega > 0$')
ax.plot(Omega.real,-Omega.imag,'b--',label=r'$L(j\omega), \omega < 0$')
ax.plot(-1.,0.,'ro',label=r'critical point $-1\in\mathbb{C}$')
ax.legend(fontsize=14,frameon=True,shadow=True)
ax.set_xlabel(r'$\operatorname{Re}\ L(j\omega)$')
ax.set_ylabel(r'$\operatorname{Im}\ L(j\omega)$');

for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(16)
```

```
k_ = np.linspace(0.1,10,70)

lambda1 = np.array([])
lambda2 = np.array([])
lambda3 = np.array([])

for k in k_ :
  num = [k]
  den = [1,2,1,k]
  tf = ctrl.TransferFunction(num,den)
  ss= ctrl.tf2ss(tf)

  eigenvalue, _ = np.linalg.eig(ss.A)
  lambda1 = np.append(lambda1, eigenvalue[0])
  lambda2 = np.append(lambda2, eigenvalue[1])
  lambda3 = np.append(lambda3, eigenvalue[2])

# confirm the non-real eigenvalues are complex-conjugates
assert np.allclose(lambda1[np.abs(lambda1.imag) > 0.],
                   np.conj(lambda2[np.abs(lambda1.imag) > 0]))

#Axis labels
plt.title('locus of eigenvalues as feedback parameter $k$ varies')
plt.xlabel(r'real part of eigenvalue ($\operatorname{Re}\ \lambda$)')
plt.ylabel(r'imaginary part of eigenvalue ($\operatorname{Im}\ \lambda$)')

#Plotting values
plt.scatter(lambda1.real,lambda1.imag,color='b')
plt.scatter(lambda2.real,lambda2.imag,color = 'r')
plt.scatter(lambda3.real,lambda3.imag,color = 'g')
# we'll only annotate every 10 of the computed eigenvalues
for k,lam1,lam2 in list(zip(k_,lambda1,lambda2))[::10]:
  plt.annotate(r'$k=%0.1f$'%k,xy=(lam1.real,lam1.imag))
  plt.annotate(r'$k=%0.1f$'%k,xy=(lam2.real,lam2.imag))
#x- and y- limits and drawing axes
ylim = plt.ylim()
plt.plot(np.zeros(2),ylim,'k--')
xlim = plt.xlim()
plt.plot(xlim,np.zeros(2),'k--')
plt.xlim(xlim); plt.ylim(ylim);
```

## subproblem (b.)

Choose a value for $k$ that results in a stable closed-loop system and determine the corresponding ***gain margin***.

```
# Choose k = 1.0
L = lambda s,k=1. : k / (s * (s+1.)**2)

omega = np.linspace(0.75,1e1,1000)
Omega = L(1.j*omega)

abs_L = np.abs(Omega)
angle_L = np.unwrap(np.angle(Omega))*180./np.pi

circle = np.exp(1.j*np.linspace(0.,2*np.pi))

# index (i.e. entry) of Omega array corresponding to stability margin
g_m_idx = (angle_L < -180).nonzero()[0][0]
phi_m_idx = np.argmin(np.min(np.abs(Omega - circle[:,np.newaxis]),axis=0))
s_m_idx = np.argmin(np.abs(Omega - (-1.)))

# numerical value of stability margin
g_m = -1./Omega[g_m_idx].real
phi_m = np.angle(Omega[phi_m_idx])*180./np.pi
phi_m_arc = np.exp(1.j*np.linspace(-180,phi_m)*np.pi/180.)
s_m = np.abs(Omega[s_m_idx] - (-1.))

axs = []

plt.figure(figsize=(6,6))
ax = plt.subplot(1,1,1); ax.grid('on'); ax.axis('equal')
# Omega, i.e. graph of L(j omega)
ax.plot(Omega.real,Omega.imag,'b-',label=r'$L(j\omega), \omega > 0$')
ax.plot(Omega.real,-Omega.imag,'b--',label=r'$L(j\omega), \omega < 0$')
# unit circle
ax.plot(circle.real,circle.imag,'k--')
# stability margins
ax.plot(-1./g_m,0.,'ro',label=r'$-1/g_m$')
ax.plot(phi_m_arc.real,phi_m_arc.imag,'r--',label=r'$\varphi_m$')
ax.plot(Omega[phi_m_idx].real,Omega[phi_m_idx].imag,'ro')
ax.plot([-1.,Omega[s_m_idx].real],[0.,Omega[s_m_idx].imag],'r.-',ms=10,label=r'$s_m$')
# critical point (-1. + 0.j)
ax.plot(-1.,0.,'ko',label=r'critical point $-1\in\mathbb{C}$')
# legend and labels
ax.legend(fontsize=14,frameon=True,shadow=True)
ax.set_xlabel(r'$\operatorname{Re}\ L(j\omega)$')
ax.set_ylabel(r'$\operatorname{Im}\ L(j\omega)$');
axs.append(ax)

for ax in axs:
  for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
               ax.get_xticklabels() + ax.get_yticklabels()):
      item.set_fontsize(16)
```

You just started a new job as a ***Senior Control System Engineer*** at a startup company that is designing a quadrotor drone for power line inspection.

## subproblem (a.)

At this morning's all-hands meeting, the mechatronics team announces that electromagnetic interference introduces significant input disturbance $v$ and output disturbance $w$ at $60\text{Hz}$, so your boss tasks you with designing controller $C$ to attenuate 90% of the effect of these disturbances on input $u$ and output $y$, i.e. with $G_{uv}$ denoting the transfer function from $v$ to $u$ and $G_{yw}$ denoting the transfer function from $w$ to $y$,

$$|G_{uv}(j\, 60\text{Hz})| \le 0.1,\ |G_{yw}(j\, 60\text{Hz})| \le 0.1. $$

Explain to your boss why it is not possible to satisfy this specification.

**Solution:**\
From the block diagram, we have:\
$y = \frac{1}{1+PC}w + \frac{P}{1+PC}v + \frac{PCF}{1+PC}r$,\
Thus, the transfer function,\
$G_{yw} = \frac{1}{1+PC}$,\
Similarly, we have the transfer function,\
$G_{uv} = \frac{PC}{1+PC}$,\
It can be seen that\
$|G_{uv}(s)| + |G_{yw}(s)| = \frac{PC}{1+PC} + \frac{1}{1+PC}  = 1$,\
Thus, it is not possible to satisfy the two requirements simultaneously.

## subproblem (b.)

During this afternoon's stand-up meeting, the aerodynamics team points out that turbulent wind conditions cause broad-band input disturbance $v$, so your boss tasks you with changing the controller design, from $C$ to $\widetilde{C}$, to decrease the impact of these disturbances on the input $u$ by 25% at all frequencies, i.e. with $G_{uv}$ denoting the transfer function from $v$ to $u$ obtained with controller $C$ and $\widetilde{G}_{uv}$ denoting the transfer function obtained with controller $\widetilde{C}$,

$$ \forall \omega \geq 0 : |\widetilde{G}_{uv}(j\,\omega)| \leq 0.75 |G_{uv}(j\,\omega)|. $$

Explain to your boss why it is not possible to satisfy this specification.

**Solution:**

The ***Bode integral formula*** ensures that the area under the log magnitude of the sensitivity function is conserved, so it is not possible to uniformly decrease the magnitude of the transfer function across all frequencies -- any decrease in one frequency range must be compensated by an increase in another.
